````python
import pandas as pd
import numpy as np
from anthropic import Anthropic
import os
````

````python
MODEL = "claude-sonnet-4-5-20250929"
client = Anthropic(api_key='')
````

````python
probs = pd.read_csv('subset_problems.csv')
probs
````

````python
langs = ['Python', 'Java', 'C']
````

````python
llm_gen = {
    'Python': [],
    'Java': [],
    'C': []
}

for i in probs.index:
    print(f'problem {i}')
    q = probs.iloc[i,1]

    for l in langs:
        prompt = f"""You are a competitive programming assistant.
Write a complete, efficient solution in {l} for the following problem:

Problem Description:
{q}

Return only the {l} code NO explanations.
Just return the RAW code.
"""
        response = client.messages.create(
            model=MODEL,
            max_tokens=2500,
            messages=[
                {"role": "user", "content": prompt}
            ]
        )

        s = response.content[0].text
        if s.startswith('```'):
            s = s.strip('```')
            s = s[s.find('\n')+1:]
        llm_gen[l].append(s)
    print(llm_gen[np.random.choice(langs)][-1])
````

````python
for l, code in llm_gen.items():
    probs[l] = code
````

````python
probs.to_csv('codenet_claude.csv',index=False)
````
